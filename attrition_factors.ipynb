```python
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

sns.set_style('ticks')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'

data = pd.read_csv("HR-Employee-Attrition.csv")
data_attr_yes = data[data.Attrition == 'Yes']
data_attr_no = data[data.Attrition == 'No']
```

```python
for i in data.columns:
    if(data[i].dtypes!=object and data[i].nunique()<30):
        print(i, data[i].unique())
```

```python
data_pie = data.Attrition.value_counts()
plt.title("Organization Attrition distribution")
plt.pie(data_pie, labels=data_pie.index, autopct='%1.1f%%')
plt.show()
```

# Observation:

Around 84% employee data is from the current employees and 16% of employees have left the organization

```python
plt.figure(figsize=(12, 10))
x = ["Low", "Medium", "High", "Very High"]
data[data['Attrition'] == 'Yes']['Education'].hist(bins=36, color='gold', label="Attrition=Yes", alpha=0.6)
plt.title("Organization Attrition Education Wise distribution")
plt.xlabel("Education")
plt.ylabel('Number of Employees left the organization')
plt.legend()
x = ["Below College", "College", "Bachelor", "Master", "Doctor"]
y_pos = [1, 2, 3, 4, 5]
plt.xticks(y_pos, x)
plt.minorticks_on()
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```

# Observation:
From the above graph, it can be found that employees with an education of Bachelors attrited the most

```python
plt.figure(figsize=(12, 10))
sns.set_style('darkgrid')
plt.title("Organization Attrition Overtime distribution")
ax=sns.countplot(data=data, x="OverTime", hue="Attrition")
ax.set_ylabel("Number of employees", fontsize = 15)
ax.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```

# Observation:
From the above graph, it can be found that employees who worked overtime attrited the most

```python
plt.figure(figsize=(12, 8))
plt.title("Organization Attriton Years Since Last Promotion Distribution")
sns.scatterplot(x=data.YearsSinceLastPromotion,y= data['YearsSinceLastPromotion'].value_counts(), hue=data.Attrition,s=100)
plt.xlabel("Years Since Last Promotion")
plt.ylabel("No of Associates")
plt.show()
```

# Observation
As we can see in the above scatter plot, all the employees who attrited have got promotion in less than 1 year duration. While employees who haven't got promotion are still with the company.

```python
bins = np.arange(20,70,5)
plt.figure(figsize=(12, 6))
plt.title('Organization Attriton Age Distribution')
plt.xlabel('Age')
plt.ylabel('Number of Employees left the organization')
plt.hist(data_attr_yes.Age, bins=bins, label='Attrition', rwidth=0.8)
plt.xticks(bins)
plt.show()
```

# Observation:
From the above graph, it can be seen that more number of employee have attrited between the age of 25-35

```python
plt.title("Gender Wise Job Satisfaction Distribution")
sns.boxplot(x=data["Gender"], y=data["JobSatisfaction"], hue=data['Attrition'], palette="Set1",
            linewidth=2.5)
plt.show()
```

# Observation
For individuals who stayed in the organization,distribution of job satisfaction is same irrespective of gender. While amongst those who left, females had lower job satisfaction.

```python
"""Job Satisfaction"""
plt.figure(figsize=(12, 12))
plt.title('Organization Attrition Job Satisfaction Wise Distribution',fontsize=15)
sns.set_style('darkgrid')
sns.countplot(x="JobSatisfaction", hue="Attrition", data=data)
plt.ylabel("No of employees")
plt.xlabel("Job Satisfaction")
plt.show()
```

```python
colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728"]
plt.title("Job involvement Distribution")
di = {1:"Very High", 2: "High", 3: "Medium", 4: "Low"}
jobInvolvement = data_attr_yes.replace({"JobInvolvement": di})
jobInvolvement["JobInvolvement"].value_counts().plot(kind='pie', autopct='%1.1f%%')
```

# Observation
From the above graph, it can be found that employees with lower job involvement attried the most

```python
plt.figure(figsize=(12, 12))
x = ["Low", "Medium", "High", "Very High"]
plt.title('Organization Attrition Environment Satisfaction Wise Distribution',fontsize=15)
plt.xlabel('Environment Satisfaction')
plt.ylabel('Number of Employees who left the organization')
plt.hist(data_attr_yes.EnvironmentSatisfaction, label='Attrition')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
y_pos = [1, 2, 3, 4]
plt.xticks(y_pos, x)
plt.show()
```

# Observation
From the above graph it can be concluded that people who were having lower environment satisfaction attired most

```python
plt.figure(figsize=(12, 6))
plt.title('Organization Distance Distribution')
plt.xlabel('Distance in Miles')
plt.ylabel('Number of Employees left the organization')
plt.hist(data_attr_yes.DistanceFromHome , bins=np.arange(0,30,3), label='Attrition');
```

# Observation 
It can be seen from plot that closeness to workplace does not ensure employee stability many of the folks leaving the organization are stay close to the company.

```python
sns.set(rc={'figure.figsize':(10,6)})
plt.title("Performance rating wise job satisfaction distribution")
sns.pointplot(y="JobSatisfaction", x="PerformanceRating", hue="Attrition", linestyles = '-.', data=data)
```

# observation
From the above graph, we see as performance increase, the job satisfaction remains less for employees who attrited.
